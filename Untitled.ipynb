```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return x**3-2*x**2-x+2

def d_func(x):
    return 3*x**2-4*x-1

```

```python
def optimize(f, x0, N=100, h=0.00000000001):
    xs = []
    d_xs = []
    for _ in range(N):
        y = f(x0)

        df_dx = (f(x0+h)-f(x0))/ h
        x1 = x0 - y/df_dx
        x0 = x1
    return xs, d_xs, x0
    
```

```python
xs, d_xs, x0 = optimize(d_func,-22, N = 10)
print(xs, d_xs, x0)
```
