```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the data
a = np.random.normal(0, 1, 1000)
b = np.random.normal(0, 0.5, 1000)
c = np.random.normal(0.5, 5, 1000)

# Plot kernel density plots
plt.figure(figsize=(10, 6))

sns.kdeplot(a, label='a (mean=0, std=1)')
sns.kdeplot(b, label='b (mean=0, std=0.5)')
sns.kdeplot(c, label='c (mean=0.1, std=1)')

plt.title('Kernel Density Plots of a, b, and c')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()

```

```python
ttest_ind(a,c)
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier
d = load_digits()
digits = d["data"]
labels = d["target"]
N = 200
idx = np.argsort(np.random.random(len(labels)))
x_test, y_test = digits[idx[:N]], labels[idx[:N]]
x_train, y_train = digits[idx[N:]], labels[idx[N:]]
clf = MLPClassifier(hidden_layer_sizes=(128,))
clf.fit(x_train, y_train)
score = clf.score(x_test, y_test)
pred = clf.predict(x_test)
err = np.where(y_test != pred)[0]
print("score      : ", score)
print("errors:")
print("  actual   : ", y_test[err])
print("  predicted: ", pred[err])
```
